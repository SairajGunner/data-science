```python
df.shape
```

```python
country_production_stats = []
```

```python
class CountryStat:
    def __init__(self, country_name, ordered_crop) -> None:
        self.country_name = country_name
        self.ordered_crop = ordered_crop
```

```python
for i in range(df.shape[0]):
    country_stat = CountryStat(df.iloc[i]["Entity"], [])
    for food in df.columns[2:]:
        country_stat.ordered_crop.append({"crop_name": food, "crop_quantity": df.iloc[i][food]})
    country_stat.ordered_crop.sort(key=lambda x: x["crop_quantity"], reverse=True)
    country_production_stats.append(country_stat)
```
